#### Decoding the Cart Column

Identifying unique heads in cart column

```python
list_column=[]

for i in data.index:
    length=len(re.sub("[^a-z0-9 A-Z:,]","",data['Cart'][i]).split(","))
    for j in range(0, length):    
        list_column.append(re.sub("[^a-z0-9A-Z:,]","",data['Cart'][i]).split(",")[j].split(":")[0])
set(list_column)
```

Forming new columns in data

```python
def new_col(data_name):
    data_name['id']=" "
    data_name['pname']=" "
    data_name['category']=" "
    data_name['quantity']=0
    data_name['price']=0
    data_name['isSubscription']= " "
    data_name['subsId']= " "


def values_input(data_name,index_1,list_index):
    first=re.sub("[^a-z0-9A-Z:,]","",data_name['Cart'][index_1])
    for i in set(list_column):   
        
        if first.split(',')[list_index].split(":")[0]==i:
            data_name[i][index_1]=" ".join(first.split(',')[list_index].split(":")[1:])
```

Code to put values in new columns 

```python
new_col(data)
lis=[]
for i in range(0,len(data)*2):
    try:
        if i in lis:
            continue

        else:
            length=len(re.sub("[^a-z0-9 A-Z:,]","",data['Cart'][i]).split(","))
            if length <=8:
                for j in range(0,length):
                    values_input(data,i,j)

            elif length > 8:  

                    for j in range(0,length):
                        values_input(data,i,j)

                    for k in range(0,int(length/5)):
                        a=int(length/5)
                        c=pd.concat([data.iloc[i:i+1]]*(a-1))
                        c.reset_index(drop=True, inplace=True)
                    new_col(c)

                    length=length-5   
                    for o in c.index:

                        for p in range(0,length):
                            values_input(c,o,p)

                        length=length-5
                    data=pd.concat([data.iloc[:i+1],c,data.iloc[i+1:]])
                    data.reset_index(drop=True, inplace=True)

                    for t in range(i+1,a+i):
                        lis.append(t)
                        
    except:
        continue
```

Applying Label Encoding On Categorical Columns

```python
df=data.copy()
cat=['City','State','isCOD','Status','IVR','Remarks','pname','category','subsId','isSubscription']
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for col in cat:
    df[col]=le.fit_transform(data[col])
```

```python
df.replace(" ",0,inplace=True)
df.fillna(0,inplace=True)
```

Selecting Best features for the Status column as output column

```python
from sklearn.feature_selection import SelectKBest  #Best Features
from sklearn.feature_selection import chi2
```

```python
x=df.drop('Status',axis=1)
y=df['Status']
```

```python
ordered_rank_feature=SelectKBest(score_func=chi2,k=20)
ordered_feature=ordered_rank_feature.fit(x,y)
ordered_feature.scores_
```

```python
data_score=pd.DataFrame(ordered_feature.scores_,columns=['score'])
```

```python
data_x=pd.DataFrame(x.columns,columns=['feature'])
features_rank=pd.concat([data_x,data_score],axis=1)
```

```python
features_rank.nlargest(12,'score') #Best Features Top 12
```
